# Data Preprocessing

columns taken: genres id keywords title overview cast crew


```python
DataFrame=df[['genres','id','keywords','title','overview','cast','crew']]
```

```python
DataFrame.head(1)
```

```python
DataFrame.isnull().sum()
```

```python
DataFrame.dropna(inplace=True)
```

```python
DataFrame.isnull().sum()
```

```python
DataFrame.iloc[0].genres
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
DataFrame['genres']=DataFrame['genres'].apply(convert)
```

```python
DataFrame.head()
```

```python
DataFrame['keywords']=DataFrame['keywords'].apply(convert)
```

```python
def convert2(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count !=3:
            L.append(i['name'])
            count=count+1
        else:
            break
    return L
```

```python
DataFrame['cast']=DataFrame['cast'].apply(convert2)
```

```python
DataFrame.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
DataFrame['crew']=DataFrame['crew'].apply(fetch_director)
```

```python
DataFrame.head()
```

```python
DataFrame['overview'][0]
```

```python
DataFrame['overview']=DataFrame['overview'].apply(lambda x:x.split())
```

```python
DataFrame.head()
```

```python
DataFrame['genres']=DataFrame['genres'].apply(lambda x:[i.replace(" ","") for i in x])
DataFrame['keywords']=DataFrame['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
DataFrame['cast']=DataFrame['cast'].apply(lambda x:[i.replace(" ","") for i in x])
DataFrame['crew']=DataFrame['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
DataFrame.head()
```

```python
DataFrame['tags'] = DataFrame['genres']+DataFrame['keywords']+DataFrame['overview']+DataFrame['cast']+DataFrame['crew']
```

```python
DataFrame.head()
```

```python
new_df = DataFrame[['id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

# Model

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
new_df[new_df['title'] == 'Spectre'].index[0]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title'] == movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:11]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
name=input("Enter a Movie Name: ")
print("\nRecommended movies are:")
recommend(name)
```
